# Analysis

### Schools with higher budgets did not yield better test results; Schools with higher spending per student under performed compared to schools with smaller budgets.
### Smaller and medium sized schools dramatically out-performed large sized schools on passing math performance.
### Charter schools out-performed the public district schools across all categories. 

```python
# Dependencies and Setup
import csv
import pandas as pd
import numpy as np
```

```python
# File to Load
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
school_data = school_data.rename(columns={'school name':'school'})
```

```python
school_data.head()
```

```python
student_data.head()
```

```python
#Combine the data into a single dataset
school_data_complete = pd.merge(school_data, student_data)
```

```python
school_data_complete.head()
```

### District Summary

```python
# Count
school_count = len(school_data_complete["school"].unique())
school_count_formatted = '{:,}'.format(school_count)
print(f"Total school count was {school_count_formatted}.")
```

```python
student_count =school_data_complete["Student ID"].count()
student_count_formatted = '{:,}'.format(student_count)
print(f"Total student count was {student_count_formatted}.")
```

```python
# Total Budget
total_budget = school_data["budget"].sum()
total_budget_formatted = '${:,.2f}'.format(total_budget)
total_budget
print(f"Total school budget was {total_budget_formatted}.")
```

```python
# Average Scores
average_math_score = school_data_complete["math_score"].mean()
average_math_score_formatted = '{:,.2f}'.format(average_math_score)
print(f"Average math score was {average_math_score_formatted}.")
```

```python
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score_formatted = '{:,.2f}'.format(average_reading_score)
print(f"Average reading score was {average_reading_score_formatted}.")
```

```python
overall_passing_rate = (average_math_score + average_reading_score)/2
overall_passing_rate_formatted = '{:,.2f}%'.format(overall_passing_rate)
print(f"Overall passing rate was {overall_passing_rate_formatted}.")
```

```python
# Passing Rates
passing_math_count = school_data_complete[(school_data_complete["math_score"] > 70)].count()["name"]
passing_math_count_formatted = '{:,}'.format(passing_math_count)
print(f"Passing math count was {passing_math_count_formatted}.")
```

```python
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage_formatted = '{:,.2f}%'.format(passing_math_percentage)
print(f"Passing math percentage was {passing_math_percentage_formatted}.")
```

```python
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] > 70)].count()["name"]
passing_reading_count_formatted = '{:,}'.format(passing_reading_count)
print(f"Passing reading count was {passing_reading_count_formatted}.")
```

```python
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage_formatted = '{:,.2f}%'.format(passing_reading_percentage)
print(f"Passing reading percentage was {passing_reading_percentage_formatted}.")
```

```python
district_summary = pd.DataFrame({"Total Schools": [school_count_formatted], 
                                 "Total Students": [student_count_formatted], 
                                 "Total Budget": [total_budget_formatted],
                                 "Average Math Score": [average_math_score_formatted], 
                                 "Average Reading Score": [average_reading_score_formatted],
                                 "% Passing Math": [passing_math_percentage_formatted],
                                 "% Passing Reading": [passing_reading_percentage_formatted],
                                 "% Overall Passing Rate": [overall_passing_rate_formatted]})

district_summary = district_summary[["Total Schools", "Total Students", "Total Budget",
                                     "Average Math Score", 
                                     "Average Reading Score",
                                     "% Passing Math",
                                     "% Passing Reading",
                                     "% Overall Passing Rate"]]

district_summary
```

### School Summary

```python
# Determine School Type
school_types = school_data.set_index(["school"])["type"]
school_types
```

```python
# Calculate student count per school
per_school_counts = school_data_complete["school"].value_counts()
per_school_counts
```

```python
# Calculate school budget and per capita spending
per_school_budget = school_data_complete.groupby(["school"]).mean()["budget"]
per_school_capita = per_school_budget / per_school_counts
per_school_capita
```

```python
# Calculate average test scores
per_school_math = school_data_complete.groupby(["school"]).mean()["math_score"]
per_school_math
```

```python
per_school_reading = school_data_complete.groupby(["school"]).mean()["reading_score"]
per_school_reading
```

```python
# Calculate passing scores
school_passing_math = school_data_complete[(school_data_complete["math_score"] > 70)]
school_passing_math.head()
```

```python
school_passing_reading = school_data_complete[(school_data_complete["reading_score"] > 70)]
school_passing_reading.head()
```

```python
per_school_passing_math = school_passing_math.groupby(["school"]).count()["name"] / per_school_counts * 100
per_school_passing_math
```

```python
per_school_passing_reading = school_passing_reading.groupby(["school"]).count()["name"] / per_school_counts * 100
per_school_passing_reading
```

```python
overall_passing_rate = (per_school_passing_math + per_school_passing_reading) / 2
overall_passing_rate
```

```python
# Convert to data frame
per_school_summary = pd.DataFrame({"School Type": school_types,
                                   "Total Students": per_school_counts,
                                   "Total School Budget": per_school_budget,
                                   "Per Student Budget": per_school_capita,
                                   "Average Math Score": per_school_math,
                                   "Average Reading Score": per_school_reading,
                                   "% Passing Math": per_school_passing_math,
                                   "% Passing Reading": per_school_passing_reading,
                                   "% Overall Passing Rate": overall_passing_rate})
```

```python
per_school_summary = per_school_summary[["School Type", "Total Students", "Total School Budget", "Per Student Budget",
                                         "Average Math Score", "Average Reading Score", 
                                         "% Passing Math", "% Passing Reading", 
                                         "% Overall Passing Rate"]]
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
```

```python
per_school_summary.head()
```

### Scores by School Spending

```python
# bins 
spending_bins = [0, 580, 610, 640, 670]
group_names = ["$0-580", "$580-610", "$610-640", "$640-670"]
```

```python
# Categorize the spending based on the bins
per_school_summary["Spending Ranges (Per Student)"] = pd.cut(per_school_capita, spending_bins, labels=group_names)
per_school_summary
```

```python
spending_math_scores = per_school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_math_scores
```

```python
spending_reading_scores = per_school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_reading_scores
```

```python
spending_passing_math = per_school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_math
```

```python
spending_passing_reading = per_school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
spending_passing_reading
```

```python
overall_passing_rate = (spending_math_scores + spending_reading_scores) / 2
overall_passing_rate
```

```python
# Assemble into data frame
spending_summary = pd.DataFrame({"Average Math Score" : spending_math_scores,
                                 "Average Reading Score": spending_reading_scores,
                                 "% Passing Math": spending_passing_math,
                                 "% Passing Reading": spending_passing_reading,
                                 "% Overall Passing Rate": overall_passing_rate})

spending_summary = spending_summary[["Average Math Score", 
                                     "Average Reading Score", 
                                     "% Passing Math", "% Passing Reading",
                                     "% Overall Passing Rate"]]
```
